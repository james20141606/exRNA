```r
m2 = m[order(mean_exp, decreasing = TRUE)[1:3000],]
```

```r
head(m2)
```

```r
Conditions = rep(1, dim(m2)[2])
Conditions
```

```r
countDeptEst <- plotCountDepth(Data = m2, Conditions = Conditions,
                               NumExpressionGroups = 10, NCores=4)
```

```r
DataNorm <- SCnorm(Data = m2, Conditions = Conditions, PrintProgressPlots = TRUE, NCores = 4)
```

```r
NormalizedData <- results(DataNorm)
```

```r
ScaleFactors <- results(DataNorm, type='ScaleFactors')
```

```r
countDepthEst.SCNORM <- plotCountDepth(Data=m2, NormalizedData = NormalizedData,
                                       Conditions = Conditions, NCores=4)
```
