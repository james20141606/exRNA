```r
filename = paste('data/expression_matrix/',filenames[1],sep = '')
savename = paste('data/normalization/tmm.',filenames[1],sep = '')
scirepcounts <- read.csv(filename, sep='\t',row.names = 1,header=TRUE)
samples_scirep <-read.csv('data/labels/scirep_binary_classes.txt',sep=',',row.names = 1,header=TRUE)
example_sce <- SingleCellExperiment(
    assays = list(counts = as.matrix(scirepcounts)), 
    colData = samples_scirep
)
keep_gene <- rowSums(counts(example_sce)) > 0
example_sce <- example_sce[keep_gene,]
## Apply TMM normalisation taking into account all genes
example_sce <- normaliseExprs(example_sce, method = "TMM")

## normalize the object using the saved size factors
example_sce <- normalize(example_sce)
```

```r
write.table(normcounts(example_sce), file=savename, sep='\t', quote=FALSE, 
            row.names=TRUE, col.names=TRUE)
```
