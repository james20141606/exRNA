```python
def to_rank(x):
    x_index = np.argsort(x)
    x_rev_index = np.empty(x.shape[0], dtype=np.int64)
    x_rev_index[x_index] = np.arange(x.shape[0], dtype=np.int64)
    x_sorted = x[x_index]
    x_rank = np.empty(x.shape, dtype=np.int64)
    x_rank[0] = 0
    x_rank[1:] = np.cumsum(~np.isclose(x_sorted[1:], x_sorted[:-1]))
    x_rank = x_rank[-1] - x_rank
    return x_rank[x_rev_index]

cpm_rank = cpm.apply(to_rank, axis=0)
cpm_rank.head()
```

```python
counts_rank = counts.apply(to_rank, axis=0)
counts_rank.head(10)
```
